Price/Area relationship of 1500 Apartments

```python
#2D Heatmap


data = pd.read_csv('https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/2_TwoNum.csv')
p = pd.DataFrame(data)
df = data
nuum = p.values
ground_living_area = [ nuum[i][0] for i in range(1460) ]
sale_price = [ nuum[i][1]/1000 for i in range(1460) ]

hres, xedges, yedges = np.histogram2d(ground_living_area, sale_price, bins=1000)


fig, ax = plt.subplots(figsize=(10, 8))


ax.set_facecolor('white')


im = ax.imshow(hres.T, extent=[xedges.min(), xedges.max(), yedges.min(), yedges.max()],
               origin='lower', cmap='viridis', aspect='auto', interpolation='nearest')


plt.colorbar(im, label='Frequency')


plt.xlabel('Ground Living Area')
plt.ylabel('Sale Price(k$)')
plt.title('2D Heatmap of Ground Living Area vs Sale Price')

plt.show()




```

```python
#Hexbin


sns.set_style("whitegrid")


sns.jointplot(x='GrLivArea', y='SalePrice', data=df, kind='hex', cmap='inferno_r')


plt.xlabel('Above Ground Living Area (sq ft)')
plt.ylabel('Sale Price ($)')
plt.title('Hexbin Plot')
plt.show()



plt.hexbin(x=df['GrLivArea'], y=df['SalePrice'], gridsize=20, cmap='inferno_r')

plt.xlabel('Above Ground Living Area (sq ft)')
plt.ylabel('Sale Price ($)')
plt.title('Hexbin Plot')

plt.colorbar(label='count')
plt.show()
```

```python
#scatterplot/scatterplot with linear regression


plt.scatter(ground_living_area, sale_price, s=35, alpha=0.5)

plt.title('scatter plot: Ground Living Area vs. Sale Price')
plt.xlabel('Ground Living Area ')
plt.ylabel('Sale Price (k$)')

plt.show()




sns.set(style="whitegrid")

sns.regplot(x='GrLivArea', y='SalePrice', data=data, scatter_kws={'alpha':0.5})

plt.xlabel('Above Ground Living Area (sq ft)')
plt.ylabel('Sale Price ($)')
plt.title('Scatterplot with Linear Regression')

plt.show()
```

```python
#histogram




plt.hist(ground_living_area, bins=50, color='blue', edgecolor='black')
plt.title('Histogram: Ground Living Area')
plt.xlabel('Ground Living Area (sqft)')
plt.ylabel('Frequency')
plt.show()

# Plotting histogram for sale_price
plt.hist(sale_price, bins=50, color='green', edgecolor='black')
plt.title('Histogram: Sale Price')
plt.xlabel('Sale Price (k$)')
plt.ylabel('Frequency')
plt.show()
```

```python
#Wordcloud




import matplotlib.pyplot as plt
from wordcloud import WordCloud
import pandas as pd


no_of_weapons = [0 if pd.isna(num) else num for num in no_of_weapons]

data_dict = dict(zip(country_name, no_of_weapons))


wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(data_dict)


plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Number of Weapons by Country')
plt.show()





data_for_plot = np.array(no_of_weapons)
data_for_plot = data_for_plot[~np.isnan(data_for_plot)]


fig, ax = plt.subplots(figsize=(12, 6))
ax.vlines(country_name, ymin=0, ymax=data_for_plot, color='blue', alpha=0.7, linewidth=2)
ax.scatter(country_name, data_for_plot, color='red', s=100, alpha=0.7)


ax.set_title('Lollipop Chart: Weapons Sales by Country')
ax.set_xlabel('Country')
ax.set_ylabel('Number of Weapons')


plt.xticks(rotation=45, ha='right')


plt.subplots_adjust(bottom=0.5)


plt.show()
```

Evolution of baby names in United States

```python
#lineplot





import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv"
babynames = pd.read_csv(url)

selected_names = ["Emma", "Madison", "Hannah", "Grace", "Emily"]
data = babynames[(babynames['Name'].isin(selected_names)) & (babynames['Gender'] == "F")]


wide = data.groupby(['Year', 'Name'])['Count'].sum().unstack()


colors = {'Emma': 'blue', 'Madison': 'orange', 'Hannah': 'green', 'Grace': 'red', 'Emily': 'purple'}


sns.set(style="whitegrid")


plt.figure(figsize=(12, 8))
sns.lineplot(data=wide, palette=colors, linewidth=2.5)

plt.title("Popularity of American names in the previous 30 years", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Count", fontsize=14)


plt.legend(title='Name', title_fontsize='14', fontsize='12', bbox_to_anchor=(1, 1), loc='upper left')

plt.tight_layout()
plt.show()







```

Perception of Probability

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from io import StringIO

# Load dataset from GitHub
url = "https://raw.githubusercontent.com/zonination/perceptions/master/probly.csv"
response = requests.get(url)
data = pd.read_csv(StringIO(response.text))

# Reshape the data
data = pd.melt(data, var_name='text', value_name='value')
data['text'] = data['text'].str.replace(".", " ")
data['value'] = round(data['value'].astype(float))

categories = ["Almost Certainly", "Very Good Chance", "We Believe", "Likely",
              "About Even", "Little Chance", "Chances Are Slight", "Almost No Chance"]

data = data[data['text'].isin(categories)]

# Set seaborn style
sns.set(style="whitegrid")

# Set color palette
colors = sns.color_palette("coolwarm", n_colors=len(categories))

# Plot
plt.figure(figsize=(16, 6))

# Violin plot
plt.subplot(1, 2, 1)
sns.violinplot(x='text', y='value', data=data, order=categories, palette=colors, inner=None)
plt.xlabel("")
plt.ylabel("Assigned Probability (%)", fontsize=12)
plt.title("Violin Plot - Probability Assignments", fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
sns.despine(left=True)

plt.legend(loc='upper right')



# Box plot
plt.subplot(1, 2, 2)
sns.boxplot(x='text', y='value', data=data, order=categories, palette=colors)
plt.xlabel("")
plt.ylabel("Assigned Probability (%)", fontsize=12)
plt.title("Box Plot - Probability Assignments", fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
sns.despine(left=True)

# Show plot
plt.tight_layout()
plt.show()
```

Life Expectency, GDP Per Capita and Population Size

```python
import plotly.express as px
import plotly.graph_objs as go
from plotly.subplots import make_subplots
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load Gapminder data for the year 2007
gapminder_data = px.data.gapminder().query("year == 2007")

# Round the GDP per capita, population, and life expectancy for better readability
gapminder_data['gdpPercap'] = round(gapminder_data['gdpPercap'], 0)
gapminder_data['pop'] = round(gapminder_data['pop'] / 1000000, 2)
gapminder_data['lifeExp'] = round(gapminder_data['lifeExp'], 1)

# Create a scatter plot with Plotly Express
fig_plotly = px.scatter(
    gapminder_data,
    x='gdpPercap',
    y='lifeExp',
    size='pop',
    color='continent',
    hover_name='country',
    log_x=True,
    size_max=60,
    color_discrete_map={'Africa': 'blue', 'Americas': 'green', 'Asia': 'red', 'Europe': 'purple', 'Oceania': 'orange'}
)
fig_plotly.update_layout(
    title='GDP per capita vs. Life expectancy in 2007',
    xaxis_title='GDP per capita (USD)',
    yaxis_title='Life expectancy (years)',
    legend_title='Continent'
)
fig_plotly.show()

# Create a matplotlib subplot with Seaborn for comparison
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 6))

# Scatter plot with GDP per capita vs population
sns.scatterplot(
    data=gapminder_data,
    x='gdpPercap',
    y='pop',
    hue='continent',
    size='lifeExp',
    sizes=(50, 1000),
    alpha=0.7,
    palette={'Africa': 'blue', 'Americas': 'green', 'Asia': 'red', 'Europe': 'purple', 'Oceania': 'orange'},
    ax=ax1
)
ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_xlabel('GDP per capita (log scale)')
ax1.set_ylabel('Population (log scale)')
ax1.set_title('GDP per capita vs population by continent')

# Scatter plot with Life expectancy vs population
sns.scatterplot(
    data=gapminder_data,
    x='lifeExp',
    y='pop',
    hue='continent',
    size='gdpPercap',
    sizes=(50, 1000),
    alpha=0.7,
    palette={'Africa': 'blue', 'Americas': 'green', 'Asia': 'red', 'Europe': 'purple', 'Oceania': 'orange'},
    ax=ax2
)
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_xlabel('Life expectancy (log scale)')
ax2.set_ylabel('Population (log scale)')
ax2.set_title('Life expectancy vs population by continent')

plt.show()
```

Comparing Raper Lyrics

```python

from matplotlib_venn import venn3


url = "https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/14_SeveralIndepLists.csv"
lyrics_df = pd.read_csv(url, delimiter=' ', skipinitialspace=True)


booba_words_set = set(lyrics_df[lyrics_df['artist'].eq('booba')]['word'])
nekfeu_words_set = set(lyrics_df[lyrics_df['artist'].eq('nekfeu')]['word'])
georges_brassens_words_set = set(lyrics_df[lyrics_df['artist'].eq('georges')]['word'])


plt.figure(figsize=(8, 8))


venn_diagram = venn3([booba_words_set, nekfeu_words_set, georges_brassens_words_set],
                     set_labels=('Booba', 'Nekfeu', 'Georges Brassens'))


venn_diagram.get_patch_by_id('100').set_color('cyan')  # Booba
venn_diagram.get_patch_by_id('010').set_color('magenta')  # Nekfeu
venn_diagram.get_patch_by_id('001').set_color('yellow')  # Georges Brassens


if venn_diagram.get_label_by_id('100') is not None:
    venn_diagram.get_label_by_id('100').set_text('Booba')
if venn_diagram.get_label_by_id('010') is not None:
    venn_diagram.get_label_by_id('010').set_text('Nekfeu')
if venn_diagram.get_label_by_id('001') is not None:
    venn_diagram.get_label_by_id('001').set_text('Georges Brassens')

plt.title("Venn Diagram of Common and Specific Words in Lyrics")
plt.show()
```
